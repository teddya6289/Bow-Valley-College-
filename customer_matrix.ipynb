```python
import pandas as pd
import numpy as np
import os,tabulate
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
def customer_model(df):

    if not df.empty:
# Features for clustering
        df.dropna(how = 'all',axis = 1)
        df.dropna(how = 'all')
        df['cust_id'] = df['cust_id'].astype(str)
        feature_df = df[['customer_retention_score','customer_profitability_score']]

        scaler = StandardScaler()

        scaled_feature_df = scaler.fit_transform(feature_df)

# Choosing the number of clusters
        kmeans = KMeans(n_clusters=2, random_state=42)
        kmeans.fit(scaled_feature_df)
        df['Cluster'] = kmeans.labels_
# THIS MODEL USES NUMBER FROM 0 TO CLUSTER DATA BASED ON A SIMILAR BEHAVIOR OF THE DATA
# NOTE: IN THIS GIVEN CASE 0 INDICATES LOYAL CUSTOMER GROUP. 1 INDICATES LESS LOYAL CUSTOMER GROUP

# Creating customer matrix to determine customer group/segment according to loyalty and profitability

        loyal_profitable_cust = df[(df['Cluster']== 0)&(df['customer_profitability_score']>0)]
        loyal_less_profitable_cust = df[(df['Cluster']== 0)&(df['customer_profitability_score']<=0)]

        less_loyal_profitable_cust = df[(df['Cluster']== 1)&(df['customer_profitability_score'] >0)]
        less_loyal_less_profitable_cust = df[(df['Cluster']== 1)&(df['customer_profitability_score'] <=0)]

# Group by cluster to understand the segment characteristics
        cluster_summary = df.groupby('Cluster').agg({'cust_id':'count',
                                                     'customer_retention_score':'mean',
                                                     'customer_profitability_score':'mean'})

        print(cluster_summary)
        print(f'LOYAL AND PROFITABLE CUSTOMER GROUP\n{tabulate.tabulate(loyal_profitable_cust.set_index('cust_id'),headers = loyal_profitable_cust.keys(),tablefmt = 'heavy_grid')}')
        print(f'LOYAL AND LESS PROFIABLE CUSTOMER GROUP\n{tabulate.tabulate(loyal_less_profitable_cust.set_index('cust_id'),headers=loyal_less_profitable_cust.keys(),tablefmt='heavy_grid')}')
        print(f'LESS LOYAL AND PROFITABLE CUSTOMER GROUP\n{tabulate.tabulate(less_loyal_profitable_cust.set_index('cust_id'),headers=less_loyal_profitable_cust.keys(),tablefmt='heavy_grid')}')
        print(f'LESS LOYAL AND LESS PROFITABLE CUSTOMER GROUP\n{tabulate.tabulate(less_loyal_less_profitable_cust.set_index('cust_id'),headers=less_loyal_less_profitable_cust.keys(),tablefmt='heavy_grid')}')
```

```python
def processdata(filename):
    path = os.path.abspath('C:/Users/Balli/Pictures/prediction/')
    fullpath = os.path.join(path,filename)
    if os.path.exists(fullpath):
        customer_ratingdataset =pd.read_csv(fullpath)
        Analyzedcustomer_ratingdata = customer_model(customer_ratingdataset)
        return Analyzedcustomer_ratingdata
```
